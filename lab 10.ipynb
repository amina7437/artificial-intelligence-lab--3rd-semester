## **Data Preprocessing:**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\SYS\\Desktop\\ai lab 10\\student_burnout_dataset (1).csv")
data
```

```python
print("Rows:", data.shape[0])
```

```python

print(data["burnout_level"].unique())
```

```python
for col in data.columns:
    data[col] = data[col].fillna(data[col].mode()[0])
```

## **Checking Null Values:**

```python

print(np.sum(pd.isnull(data)))
```

```python
print(data.isnull().sum())
```

## **Checking Data Types:**

```python
print(data.dtypes)
```

```python
categorical_columns = data.select_dtypes(include=['object', 'string']).columns
print(categorical_columns)
```

## **Converting Data Types:**


```python

for col in categorical_columns:
    data[col] = pd.factorize(data[col])[0]
```

```python
float_columns = data.select_dtypes(include=['float64']).columns
print(float_columns)
```

```python
for col in float_columns:
    data[col] = data[col].astype(int)
```
